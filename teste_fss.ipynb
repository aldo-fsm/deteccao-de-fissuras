```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from optimization.fss import FSS
import holoviews as hv
%matplotlib inline
# equivalent to plt.rcParams['animation.html'] = 'html5'
plt.rc('animation', html='jshtml')
```

```python
np.seterr(all='raise')
```

```python
def fitness(x):
    return -np.sum(x**2)
@np.vectorize
def aux(x, y):
    return -(x**2+y**2)
def min_max(x):
    return np.min(x), np.max(x)
```

```python
optimizer = FSS(10, 0.1, 0.01)
aquarium = np.array([[0,1],[0,1]])
optimizer.initSwarm(100, aquarium, 100)
```

```python
def plot(data):
    return hv.Curve(data)
```

```python
buffer = hv.streams.Buffer([])
dm = hv.DynamicMap(plot, streams=)
```

```python
data = []
for i in range(optimizer.iterations):
    optimizer.step(fitness)
    #data.append(optimizer.X)
    #print(*optimizer.best)
    print('========================')
```

```python
from matplotlib.animation import FuncAnimation
%matplotlib inline
```

```python
%%capture
fig, ax = plt.subplots()
pop = data[0]
X, Y = np.meshgrid(np.linspace(aquarium[0,0], aquarium[0,1]), np.linspace(aquarium[1,0], aquarium[1,1]))
Z = aux(X,Y)
pcolor = ax.pcolormesh(X, Y, Z);
scatter, = ax.plot(*zip(*pop), marker='o', ls='');

def animate(i):
    if len(data) <= i+1:
        print('acabou')
        return scatter, pcolor
    pop = data[i+1]
    #print(i, np.linalg.norm(pop))
    scatter.set_data(*list(zip(*pop)))
    return scatter, pcolor
ax.set_xlim([0,1])
ax.set_ylim([0,1])
anim = FuncAnimation(fig, animate, frames=len(data)-1, blit=True, interval=100);
```
